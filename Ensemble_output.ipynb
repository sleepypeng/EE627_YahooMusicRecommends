```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('Result6(PCA3).xlsx', index_col=0)
```

```python
predictDic = {}
for row in data.iterrows():
    ID = row[0]
    score = row[1]
    score = float(score)
    ID1 = ID.strip().split('_')
    user = ID1[0]
    item = ID1[1]
    predictDic.setdefault(user, {})[item] = score
```

```python
predictDic
```

```python
for key in predictDic:
    predictDic[key] = sorted(predictDic[key].items(), key = lambda x : x[1], reverse = True)
predictDic
```

```python
item_score = []
for key in predictDic:
    for i in range(6):
        if i < 3:
            need1 = key + '_' + predictDic[key][i][0] + '|' + str(1)
            item_score.append(need1)
        else:
            need2 = key + '_' + predictDic[key][i][0] + '|' + str(0)
            item_score.append(need2)
```

```python
len(item_score)
```

```python
result_list = []
for item in item_score:
    line = item.strip().split('|')
    ID = line[0]
    score = line[1]
    result = [ID, score]
    result_list.append(result)
name = ['TrackID', 'Predictor']
test=pd.DataFrame(columns=name, data=result_list)
test.to_csv('FinalResult6(PCA3).csv',encoding='gbk')
```
